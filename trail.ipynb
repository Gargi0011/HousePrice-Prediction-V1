```python
def handling_missing_values(train):
    """
    Handle missing values in the train and test datasets.
    Fill numerical columns with median and categorical columns with mode.
    """
    # Calculate the ratio of missing values for each column

    missing_ratio = train.isnull().sum() / len(train)

    # Filter columns with missing values
    missing_cols = missing_ratio[missing_ratio > 0].index

    # Fill missing values
    for col in missing_cols:
        if train[col].dtype in ['int64', 'float64']:
            # Fill numerical columns with median
            train[col].fillna(train[col].median(), inplace=True)
        else:
            # Fill categorical columns with mode
            train[col].fillna(train[col].mode()[0], inplace=True)
```

```python
def get_skewed_columns(df, threshold=0.5, dropna=True):
    """
    Returns a sorted Series of skewed numerical features above a given threshold.

    Parameters:
    - df (pd.DataFrame): The input dataframe.
    - threshold (float): Absolute skewness above which a column is considered skewed.
    - dropna (bool): Whether to drop NA values before computing skewness.

    Returns:
    - pd.Series: Sorted skewed features and their skewness values.
    """
    # Select numeric columns
    numeric_cols = df.select_dtypes(include=['int64', 'float64'])
    
    # Compute skewness
    if dropna:
        skewness = numeric_cols.apply(lambda x: x.dropna().skew())
    else:
        skewness = numeric_cols.apply(lambda x: x.skew())
        
    # Filter based on threshold
    skewed = skewness[abs(skewness) > threshold].sort_values(ascending=False)
    
    return skewed
```

```python
from sklearn.preprocessing import PowerTransformer

def fix_skewed_features(df, threshold=0.75, method='yeo-johnson'):
    """
    Detects and transforms skewed numerical features in the DataFrame.

    Parameters:
    - df (pd.DataFrame): Input DataFrame.
    - threshold (float): Absolute skewness threshold for transformation.
    - method (str): Transformation method - 'log' or 'yeo-johnson'.

    Returns:
    - df_transformed (pd.DataFrame): DataFrame with transformed features.
    - skewed_feats (pd.Series): Skewed features and their original skew values.
    """
    df = df.copy()
    numeric_cols = df.select_dtypes(include=['int64', 'float64'])
    
    # Compute skewness
    skewness = numeric_cols.apply(lambda x: x.dropna().skew())
    skewed_feats = skewness[abs(skewness) > threshold]
    
    if method == 'log':
        for col in skewed_feats.index:
            if (df[col] >= 0).all():  # log1p requires non-negative
                df[col] = np.log1p(df[col])
    elif method == 'yeo-johnson':
        pt = PowerTransformer(method='yeo-johnson')
        df[skewed_feats.index] = pt.fit_transform(df[skewed_feats.index])
    else:
        raise ValueError("Method must be either 'log' or 'yeo-johnson'")
    
    return df, skewed_feats
```

```python
# Feature Creation
df['TotalSF'] = df['TotalBsmtSF'] + df['1stFlrSF'] + df['2ndFlrSF']
df['TotalBath'] = df['FullBath'] + (0.5 * df['HalfBath']) + df['BsmtFullBath'] + (0.5 * df['BsmtHalfBath'])
df['HouseAge'] = df['YrSold'] - df['YearBuilt']
df['RemodAge'] = df['YrSold'] - df['YearRemodAdd']
df['GarageAge'] = df['YrSold'] - df['GarageYrBlt']
df['TotalPorchSF'] = df['OpenPorchSF'] + df['EnclosedPorch'] + df['3SsnPorch'] + df['ScreenPorch']
df['HasPool'] = df['PoolArea'].apply(lambda x: 1 if x > 0 else 0)
df['HasFireplace'] = df['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
df['HasGarage'] = df['GarageArea'].apply(lambda x: 1 if x > 0 else 0)
df['HasBsmt'] = df['TotalBsmtSF'].apply(lambda x: 1 if x > 0 else 0)
```

```python
from sklearn.feature_selection import VarianceThreshold

def remove_low_variance(df, threshold=0.0):
    selector = VarianceThreshold(threshold)
    numeric = df.select_dtypes(include=['int64', 'float64']).drop(columns=['Id', 'SalePrice'], errors='ignore')
    selector.fit(numeric)
    selected_columns = numeric.columns[selector.get_support()]
    return list(selected_columns)
```

```python
def remove_highly_correlated(df, threshold=0.9):
    numeric = df.select_dtypes(include=['int64', 'float64']).drop(columns=['Id', 'SalePrice'], errors='ignore')
    corr_matrix = numeric.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [column for column in upper.columns if any(upper[column] > threshold)]
    return [col for col in numeric.columns if col not in to_drop]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
def model_based_selection(df, target_col='SalePrice', top_k=20):
    df_copy = df.copy()
    # Label encode object cols
    for col in df_copy.select_dtypes(include='object'):
        df_copy[col] = df_copy[col].astype(str).fillna('None')
        df_copy[col] = LabelEncoder().fit_transform(df_copy[col])

    X = df_copy.drop([target_col, 'Id'], axis=1, errors='ignore')
    y = df_copy[target_col]

    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X, y)

    importances = pd.Series(model.feature_importances_, index=X.columns)
    return list(importances.sort_values(ascending=False).head(top_k).index)
```

```python
# 1. Step-by-step selection
low_var_feats = remove_low_variance(df)
df_low_var = df[low_var_feats + ['SalePrice']]

# 2. Remove high correlations
non_corr_feats = remove_highly_correlated(df_low_var)
df_non_corr = df[non_corr_feats + ['SalePrice']]

# 3. Model-based top K features
selected_feats = model_based_selection(df_non_corr, top_k=25)

# Final feature set
df_final = df[selected_feats + ['SalePrice']]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor

def build_model_pipeline(df, target_col='SalePrice', model=None):
    # Default model
    if model is None:
        model = RandomForestRegressor(n_estimators=100, random_state=42)

    # Split features
    X = df.drop([target_col], axis=1)
    y = df[target_col]

    # Identify column types
    num_cols = X.select_dtypes(include=['int64', 'float64']).drop(columns=['Id'], errors='ignore').columns.tolist()
    cat_cols = X.select_dtypes(include='object').columns.tolist()

    # Define transformers
    num_transformer = StandardScaler()
    cat_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

    # Create preprocessor
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', num_transformer, num_cols),
            ('cat', cat_transformer, cat_cols)
        ],
        remainder='drop'
    )

    # Create full pipeline
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', model)
    ])

    return pipeline, X, y
```

```python
pipeline, X, y = build_model_pipeline(df_final)
pipeline.fit(X, y)

# Predictions
y_pred = pipeline.predict(X)
```

```python
def pandas_to_pydantic_dtypes(df):
    dtype_mapping = {
        'int64': 'int',
        'int32': 'int',
        'float64': 'float',
        'float32': 'float',
        'object': 'str',
        'bool': 'bool',
        'datetime64[ns]': 'datetime'
    }

    return {
        col: dtype_mapping.get(str(dtype), 'str')  # default to str if unknown
        for col, dtype in df.dtypes.items()
    }
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_model_performance(model, X, y, dataset_name="Train"):
    y_pred = model.predict(X)

    r2 = r2_score(y, y_pred)
    mae = mean_absolute_error(y, y_pred)
    rmse = np.sqrt(mean_squared_error(y, y_pred))

    print(f"ðŸ“Š {dataset_name} Set Performance:")
    print(f"RÂ² Score: {r2:.4f}")
    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    
    return {"r2": r2, "mae": mae, "rmse": rmse}
```

```python
# After fitting the pipeline
pipeline.fit(X, y)

# Evaluate on training set (or use train_test_split for better validation)
evaluate_model_performance(pipeline, X, y)
```
