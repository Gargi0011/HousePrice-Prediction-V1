```python
train=pd.read_csv('data/train.csv')
test=pd.read_csv('data/test.csv')
```

```python
threshold = 0.3  # 30%
missing_ratio = train.isnull().sum() / len(train)
cols_over_30pct = missing_ratio[missing_ratio > threshold].sort_values(ascending=False)
print(cols_over_30pct)
```

```python
train.drop(columns=cols_over_30pct.index, inplace=True)
test.drop(columns=cols_over_30pct.index, inplace=True)
```

```python
train.shape
```

```python
train.isnull().sum().sort_values(ascending=False)
```

```python
def handling_missing_values(train):
    """
    Handle missing values in the train and test datasets.
    Fill numerical columns with median and categorical columns with mode.
    """
    # Calculate the ratio of missing values for each column

    missing_ratio = train.isnull().sum() / len(train)

    # Filter columns with missing values
    missing_cols = missing_ratio[missing_ratio > 0].index

    # Fill missing values
    for col in missing_cols:
        if train[col].dtype in ['int64', 'float64']:
            # Fill numerical columns with median
            train[col].fillna(train[col].median(), inplace=True)
        else:
            # Fill categorical columns with mode
            train[col].fillna(train[col].mode()[0], inplace=True)
```

```python
handling_missing_values(train)
handling_missing_values(test)
```
